## Predict the percentage of an student based on the no. of study hours.

### DATASET : https://bit.ly/3kXTdox

### Import_libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv")
df.head()
```

### Data_Visualization Using Matplotlib.pyplot

```python
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.title('Hours vs Scores')
plt.scatter(df['Hours'],df['Scores'])
plt.show()
```

### Data Preparation Separating input_variable as X output_variable as y

```python
X = df.iloc[:,:-1].values
y = df.iloc[:,1].values
```

### Splitting into two sets train and test

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
len(X_train)
```

```python
len(X_test)
```

### Creating a Model to train the set

#### Here I used a linear regression to predict the score based on the study hours 

#### Training a model

```python
model = LinearRegression()
model.fit(X_train ,y_train)
```

### Making predictions

```python
model.predict([[9.25]])
```

### Accuracy of the model

```python
model.score(X_test,y_test)
```

### Linear regression between the test preparation

```python
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.title('Hours vs Scores')
plt.scatter(df['Hours'],df['Scores'])
plt.plot(df['Hours'],model.predict(df[['Hours']]))
plt.show()
```

```python
y_predict = model.predict(X_test)
```

### Difference between the Actual Output vs Predicted Output

```python
Diff = pd.DataFrame({'Actual Output':y_test,'Predicted':y_predict})
Diff
```
